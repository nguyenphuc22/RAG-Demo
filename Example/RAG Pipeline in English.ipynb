```python
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L12-v2")
def get_embedding(text: str) -> list[float]:
    if not text.strip():
        print("Attempted to get embedding for empty text.")
        return []

    embedding = embedding_model.encode(text)

    return embedding.tolist()
```

```python
def vector_search(user_query, collection, limit=4):
    """
    Perform a vector search in the MongoDB collection based on the user query.

    Args:
    user_query (str): The user's query string.
    collection (MongoCollection): The MongoDB collection to search.

    Returns:
    list: A list of matching documents.
    """

    # Generate embedding for the user query
    query_embedding = get_embedding(user_query)

    if query_embedding is None:
        return "Invalid query or embedding generation failed."

    # Define the vector search pipeline
    vector_search_stage = {
        "$vectorSearch": {
            "index": "vector_index",
            "queryVector": query_embedding,
            "path": "embedding",
            "numCandidates": 400,
            "limit": limit,
        }
    }

    unset_stage = {
        "$unset": "embedding"
    }

    project_stage = {
        "$project": {
            "_id": 0,
            "name": 1,
            "description": 1,
            "price": 1,
            "rating": 1,
            "score": {
                "$meta": "vectorSearchScore"
            }
        }
    }

    pipeline = [vector_search_stage, unset_stage, project_stage]

    # Execute the search
    results = collection.aggregate(pipeline)

    return list(results)
```

```python
def get_search_result(query, collection):
    get_knowledge = vector_search(query, collection, 10)
    search_result = ""
    i = 0
    for result in get_knowledge:
        print(result)
        if result.get('name'):
            i += 1
            search_result += f"\n {i}) Name: {result.get('name')}"

            if result.get('price'):
                search_result += f", Price: {result.get('price')}"
            else:
                # Mock up data
                # Retrieval model pricing from the internet.
                search_result += f", Price: Contact for more information"

            if result.get('rating'):
                search_result += f", Rating: {result.get('rating')}"

    return search_result
```

```python
# Conduct query with retrival of sources
query = "please give me price of Samsung Galaxy 5G, 4G! Why i should by this?"
source_information = get_search_result(query.lower(), collection)
source_information
```

```python
combined_information = f"Become a sales consultant for a phone store. Customer's question: {query}\nAnswer the question based on the following product information: {source_information}."
print('==== LLMs Inference')
print(combined_information)
```
