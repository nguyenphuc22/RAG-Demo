```python
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('/content/extracted_data.csv')

print(df)  # Print the first few rows to check
```

```python
df = df.rename(columns={'priceInfo.linePrice': 'price'})
```

```python
df.head()
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L12-v2")
def get_embedding(text: str) -> list[float]:
    if not text.strip():
        print("Attempted to get embedding for empty text.")
        return []

    embedding = embedding_model.encode(text)

    return embedding.tolist()
```

```python
get_embedding("Ngoc is speaking good?")
```

```python
df["embedding"] = df["name"].apply(get_embedding)

df.head()
```

```python
collection = db['embedding_for_vector_search']
```

```python
documents = df.to_dict("records")
collection.insert_many(documents)

print("Data ingestion into MongoDB completed")
```
